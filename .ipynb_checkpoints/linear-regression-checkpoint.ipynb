```python
import numpy as n


def generate_set(max, generator):
    input = []
    for i in range(0, max):
        input.append((i, generator(i)))
    return n.array(input)

def generate_3_set(max, generator):
    input = []
    for i in range(0, max):
        for j in range(0, max):
            input.append((i, j, generator(i, j)))
    return n.array(input)

def generate_4_set(max, generator):
    input = []
    for i in range(0, max):
        for j in range(0, max):
            for k in range(0, max):
                input.append((i, j, k, generator(i, j, k)))
    return n.array(input)


def lr_1_param():
    generator = lambda x : 5*x
    a = generate_set(1000, generator)
    point = a[5]
    print point
    y = point[1]
    x = point[0]
    t = (1.0/(x*x))*x*y
    print t
    

def lr_2_param():
    generator = lambda x : 7*x + 11
    values = generate_set(5, generator)
    x = [[x[0], 1] for x in values]
    x = n.matrix(x)
    y = values[:,1]
    y = n.matrix(y)
    xt = x.transpose()
    tp = xt * x
    itp = n.linalg.inv(tp)
    titp = itp * xt
    
    # Use the transpose of Y here; see defn of y in 3.16
    t = titp * y.transpose()
    print t
    
    
def lr_3_param():
    generator = lambda x, y : 7*x + 12*y + 11
    values = generate_3_set(5, generator)
    x = [[a[0], a[1], 1] for a in values]
    x = n.matrix(x)
    y = [a[2] for a in values]
    y = n.matrix(y)
    xt = x.transpose()
    tp = xt * x
    itp = n.linalg.inv(tp)
    titp = itp * xt
    
    # Use the transpose of Y here; see defn of y in 3.16
    t = titp * y.transpose()
    print t
    
    
def lr_4_param():
    generator = lambda x, y, z : 7*x + 12*y + 5*z + 11
    values = generate_4_set(5, generator)
    x = [[a[0], a[1], a[2], 1] for a in values]
    x = n.matrix(x)
    y = [a[3] for a in values]
    y = n.matrix(y)
    xt = x.transpose()
    tp = xt * x
    itp = n.linalg.inv(tp)
    titp = itp * xt
    
    # Use the transpose of Y here; see defn of y in 3.16
    t = titp * y.transpose()
    print t
    
    
def lr_hd_2_param():
    # We use a quadratic here
    generator = lambda x : 7*x**2 + 11
    values = generate_set(5, generator)
    
    # because we have a quadratic, we square
    # the x value in the X value matrix below
    # to keep the solution linear in the higher
    # dimentional space
    x = [[x[0]**2, 1] for x in values]
    x = n.matrix(x)
    y = values[:,1]
    y = n.matrix(y)
    xt = x.transpose()
    tp = xt * x
    itp = n.linalg.inv(tp)
    titp = itp * xt
    
    # Use the transpose of Y here; see defn of y in 3.16
    t = titp * y.transpose()
    print t
    
    
lr_hd_2_param()
```
